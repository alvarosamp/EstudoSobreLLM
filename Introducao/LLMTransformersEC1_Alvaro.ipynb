## Analise de Sentimentos

Análise de sentimentos é um técnica de processamento de linguagem natural que visa identificar e classificar emoçoes, opiniões ou atitutdes em relação a um determinado assunto , conteudo ou objeto. Isso é feito com base em uma coleção de dados previamente classificados, como textos, audisos ou videos, e aplicando algoritmos de aprendizado de maquina para detectar padroes e classificar novos dados. A analise de sentimentos é amplamente utilizada em diversas aplicações, como análise de midia social, avaliação de produtos serviçoes, e monitoramento de opiniões publicas.

### Classificação Zero-Shot

Classificação Zero-Shot é uma técnica de aprendizado de máquina que permite classificar objetos em categorias sem ter visto exemplos de treinamento para essas categorias específicas. Em outras palavras, ao invés de treinar um modelo de aprendizado de máquina com dados previamente rotulados para cada categoria, a Classificação Zero-Shot usa informações semânticas ou descrições das categorias, como atributos ou palavras-chave, para realizar a classificação. Essa técnica é particularmente útil em casos em que existem muitas categorias ou quando é necessário lidar com novas categorias que não foram previamente vistas durante o treinamento do modelo.

Modelo Pré-Treinado:

https://huggingface.co/facebook/bart-large-mnli

```python

modelo_2 = "facebook/bart-large-mnli"
```

```python
classificador = pipeline("zero-shot-classification", model=modelo_2, tokenizer=modelo_2)
```

```python
#Executando o classificador
resultado_zero_shot = classificador(
    "Curso sobre transformers",
    candidate_labels=['curso', 'transformers', 'machine learning', 'deep learning']
)
print(resultado_zero_shot)
```

A geração de texto pode ser usada em aplicaçãoes como chabots, assistentes virtuais, geração automatica de noticias ou resumos, entre outras. Alguns dos desafios enfrenteados na gera~çao de texto inclui a produção de texto coerente, a geração de teo relevante e a gerção de texto que siga padroes de esririta e estilo especifico

### Named Entity recoignition (classificação de textos)


reconhecimento de Entidade Nomeadas (NER), tambem conhecidso como classidficador de token, é uma tarefa em NLP onde as entridades nomeadas como pessoas , organizações , locais. O objetivo do NER é extrair informações esdtrturadas a partir de dados de texto nao estruturados e categorizar as entidades em categorias pré-definidas, como pessoa, local, oragnização. Sistemas de NER usam ténicas como métodos baseados em regra, métodos baseados em dicionários e métodos baseados em aprendizado de maquina para realizaar o reconheciment oe classficalção de entidades nomeadas .O Ner é uma etapa importante em muitas aplicações de PNL, incluindo recuperação de informações, resposta a pergunta e extração de eventos.

### Perguntas  e respostas

A resposta a perguntas(Question-answering -QA). O objetivo desses sistemas é entender a qeustao, recuperar informações relevantes de uma fonte, como um banco de dados ou em um documento, e gerar uma resposta na forme de uma resposta de texto. Esses sitemas usam varias técnicas de PLN, como recuperação de inforçãos, resumo do texto, reconhecimento de entidade nomeada e aprendizado de maquina para extrair informações relevantes e gerar uma resposta. A respoista a pergunta é uma tarefa desafiadora em PLN, pois requer a capacidade de enteder questoes complexas de linguagem naturalk e recurar informçoes relevantes de um grande corpus de dados de texto. Esses sistmas sao usados em varias aplicações, incluidno atendimento ao cliente, gernciamento de conheciemnto e tecnologia edcucacional
