## Analise de Sentimentos

Anlise de sentimentos  um tcnica de processamento de linguagem natural que visa identificar e classificar emooes, opinies ou atitutdes em relao a um determinado assunto , conteudo ou objeto. Isso  feito com base em uma coleo de dados previamente classificados, como textos, audisos ou videos, e aplicando algoritmos de aprendizado de maquina para detectar padroes e classificar novos dados. A analise de sentimentos  amplamente utilizada em diversas aplicaes, como anlise de midia social, avaliao de produtos servioes, e monitoramento de opinies publicas. 

Modelo pr treinado utilizado : 
https://huggingface.co/citizenlab/twitter-xlm-roberta-base-sentiment-finetunned


```python
# Define o modelo a ser usado
modelo_1 = "citizenlab/twitter-xlm-roberta-base-sentiment-finetunned"
```

```python
#Cria o classifcador
classificador = pipeline("sentiment-analysis", model=modelo_1, tokenizer=modelo_1)
#Aplica o classificador
```

### Classificao Zero-Shot 

Classificao Zero-Shot  uma tcnica de aprendizado de mquina que permite classificar objetos em categorias sem ter visto exemplos de treinamento para essas categorias especficas. Em outras palavras, ao invs de treinar um modelo de aprendizado de mquina com dados previamente rotulados para cada categoria, a Classificao Zero-Shot usa informaes semnticas ou descries das categorias, como atributos ou palavras-chave, para realizar a classificao. Essa tcnica  particularmente til em casos em que existem muitas categorias ou quando  necessrio lidar com novas categorias que no foram previamente vistas durante o treinamento do modelo.

Modelo Pr-Treinado:

https://huggingface.co/facebook/bart-large-mnli

```python

modelo_2 = "facebook/bart-large-mnli"
```

```python
!pip install hf_xet
```

```python
classificador = pipeline("zero-shot-classification", model=modelo_2, tokenizer=modelo_2)
```

```python
#Executando o classificador
resultado_zero_shot = classificador(
    "Curso sobre transformers",
    candidate_labels=['curso', 'transformers', 'machine learning', 'deep learning']
)
print(resultado_zero_shot)
```

```python
#Executa o classifciador com as possiveis categorias 
classificador([
    'this year , a new law on animal rights was enacted.',
    'The world cup will was held int qatar in 2022',
    'I prefer the songs from the 90s compared to the songs today.'],
    candidate_labels=['politics', 'sports', 'music', 'entertainment','forest']
    )
```
